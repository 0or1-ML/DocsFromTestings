```python
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt
from matplotlib import rc
```

```python
''' 1. CLEANING '''
```

```python
'''
Questions: 
- How many rows have null values (if there is a null value, thatâ€™s also counted)?
- How many orders have been made (not including nulls)?
'''
```

```python
df = pd.read_csv('../../tech-purchase.csv')
amount_of_nans = 0
```

```python
# cleaning dataset from mistakes.
df = df.drop(df.loc[df['Order ID'] == 'Order ID'].index)

# "for loop" to count number of Null values.
for index, row in df.iterrows():
    if pd.isnull(row['Order ID']):
        amount_of_nans += 1

print('Total number of orders:', df.shape[0])
print('Number of "Null" rows:', amount_of_nans)
print('Number of actual orders:', (df.shape[0] - amount_of_nans))
```

```python
'''
Tasks:
a) Show all of the null containing rows.
b) Drop rows that contain any null values.
'''
```

```python
# showing all null containing rows.
for index, row in df.iterrows():
    if pd.isnull(row['Order ID']):
       print(row)
```

```python
# droping the raws containing null values.
df = df.dropna()

df = df.reset_index(drop=True)

# saving cleaned data.
df.to_csv('./tech-purchase-clean1.csv', index=False)
```

```python
''' 2. RESTRUCTURING THE DATASET '''


'''
Tasks:
a) Using our (cleaned) dataset, create a new column that would contain only the months from the Order Date column.
b) Convert the month numbers into characters (f.e: 1 into January, 2 into February, and so on).
c) Create a new column that would contain only the state names taken from the Purchase Address column.
d) Convert the Quantity Ordered columnâ€™s values into an integer type and the Price Each columnâ€™s values into a floating type.
'''
```

```python
df = pd.read_csv('./tech-purchase-clean1.csv')

order_date = list()
formatted_order_date = list()
address = list()
formatted_address = list()

# Getting original "Order Date" and "Purchase Address".
for index, row in df.iterrows():
    order_date.append(df.iloc[index, 4])
    address.append(df.iloc[index, 5])

# Convertion of original "Order Date" to month name and
# original "Purchase Address" to postal abbreviation.
for i in range(df.shape[0]):
    string = datetime.strptime(order_date[i], '%m/%d/%y %H:%M').strftime('%B')
    formatted_order_date.append(string)
    string = address[i].split(', ')[-1][:2]
    formatted_address.append(string)

# Adding "Order Date Month", adding "State Abbreviation" column,
# converting "Quantity Ordered" and "Price Each" to int and float respectively.
for i in range(df.shape[0]):
    df.at[i, 'Order Date Month'] = formatted_order_date[i]
    df.at[i, 'State Postal Abbreviation'] = formatted_address[i]
    df.at[i, 'Quantity Ordered'] = int(df.iloc[i, 2])
    df.at[i, 'Price Each'] = float(df.iloc[i, 3])

# Saving to a new csv.
df.to_csv('./tech-purchase-clean2.csv', index=False)
```

```python
''' 3. GET THE USEFUL INFORMATION '''


'''
Questions:
1. Which month has the highest number of sales (in amounts)? And how much was that?
2. Plot a bar plot that would represent the sales info for each month.
3. Which state has ordered the least and which the highest?
4. At what time of the day the highest amount of money is spent on sales?
'''
```

```python
df = pd.read_csv('./tech-purchase-clean2.csv')

# Question 1.
max = df['Quantity Ordered'].max()
print(df.loc[df['Quantity Ordered'] == max])
# Expected Answer:
# June and July have the highest amount of sales of 9.
```

```python
# Question 2.
months= ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]
# Sorting rows by months.
df['Order Date Month'] = pd.Categorical(df['Order Date Month'], categories=months, ordered=True)
df.sort_values(by='Order Date Month', ascending=True, inplace=True)
df.to_csv('./tech-purchase-clean3.csv', index=False)

# Gettin total qunatities ordered by months.
months_and_quantites_ordered = df.groupby(['Order Date Month'])['Quantity Ordered'].sum()
quantities_ordered = [i for i in months_and_quantites_ordered]

# Plotting a bar chart.
yticks = [0, 2000, 5000, 7000, 10000, 15000, 20000, 22000, 25000, 27000, 30000, 10903, 13449, 17005, 20558, 18667, 15253, 16072, 13448, 13109, 22703, 19798, 28114]

rc('ytick', labelsize=7)
rc('xtick', labelsize=10)
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 8))
plt.title('Quantities Ordered by Months')
plt.bar(months, quantities_ordered, width=0.5)
plt.ylabel('Number of Ordered Products')
plt.yticks(yticks)

plt.savefig('./barchart-c3-q2.png')
```

```python
# Question 3.
states_min = set()
states_max = set()

max = df['Quantity Ordered'].max()
min = df['Quantity Ordered'].min()

# Location States that ordered max and min amounts.
for index, row in df.iterrows():
    if df.iloc[index, 2] == min:
        states_min.add(df.iloc[index, -1])
    elif df.iloc[index, 2] == max:
        states_max.add(df.iloc[index, -1])

print('States that ordered the highes amount: ', states_max)
print('States that ordered the least amount: ', states_min)
```

```python
# Question 4.
order_date = list()
order_time = list()
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

df = pd.read_csv('./tech-purchase-clean3.csv')

# Creating column 'Total Sales'.
df['Total Sales'] = df['Quantity Ordered'] * df['Price Each']

# Locating 'Order Date'.
for index, row in df.iterrows():
    order_date.append(df.iloc[index, 4])

# Extracting time from 'Order Date'.
for i in range(df.shape[0]):
    string = datetime.strptime(order_date[i], '%m/%d/%y %H:%M').strftime('%H:%M')
    order_time.append(string)

# Creating column 'Order Time'.
for i in range(df.shape[0]):
    df.at[i, 'Order Time'] = order_time[i]

del order_time
order_time = list()

# Finding max sales for each month.
max_sales_by_months = df.groupby(['Order Date Month'], sort=False)['Total Sales'].max()
max_sales_by_months = [i for i in max_sales_by_months]

# Getting time for each month's max total sales.
for i in range(len(months)):
    string = df.loc[(df['Order Date Month'] == months[i]) & (df['Total Sales'] == max_sales_by_months[i])]
    string = string.iloc[0, -1]
    order_time.append(string)

# Printing out month's time of the day highest sale.
for i in range(len(months)):
    print(f'Highest amount of sales in {months[i]} of ${max_sales_by_months[i]} is at {order_time[i]} time.')
```

```python
# Task 1.
df = pd.read_csv('./tech-purchase-clean3.csv')
items = df['Product'].unique()
items = list(items)
items_formatted = list()
quantity_of_items = list()

quantity = df.groupby(['Product'], sort=False)['Quantity Ordered'].sum()
quantity_of_items = [i for i in quantity]

for i in range(len(items)):
    product = items[i]
    product = product.replace(' ', '\n')
    items_formatted.append(product)

rc('ytick', labelsize=7)
rc('xtick', labelsize=6)
plt.figure(figsize=(12, 8))
plt.title('Total Number of Product Orders')
plt.bar(items_formatted, quantity_of_items, width=0.5)
plt.savefig('./barchart-c4-q1.png')
```

```python
# Task 1.
df = pd.read_csv('./tech-purchase-clean3.csv')
months = df['Order Date Month'].unique()
order_date = list()
order_time = list()

# Creating column 'Total Sales'.
df['Total Sales'] = df['Quantity Ordered'] * df['Price Each']

# Locating 'Order Date'.
for index, row in df.iterrows():
    order_date.append(df.iloc[index, 4])

# Extracting time from 'Order Date'.
for i in range(df.shape[0]):
    string = datetime.strptime(order_date[i], '%m/%d/%y %H:%M').strftime('%H:%M')
    order_time.append(string)

# Creating column 'Order Time'.
for i in range(df.shape[0]):
    df.at[i, 'Order Time'] = order_time[i]

del order_time
order_time = list()

# Finding max sales for each month.
max_sales_by_months = df.groupby(['Order Date Month'], sort=False)['Total Sales'].max()
max_sales_by_months = [i for i in max_sales_by_months]

# Getting time for each month's max total sales.
for i in range(len(months)):
    string = df.loc[(df['Order Date Month'] == months[i]) & (df['Total Sales'] == max_sales_by_months[i])]
    string = string.iloc[0, -1]
    order_time.append(string)

# Sorting time stamps.
sorted_time = sorted(order_time)

rc('xtick', labelsize=6)
plt.figure(figsize=(9, 7))
plt.title('Time of the Day of the Highest Amount of Sales by Month')
plt.scatter(months, order_time)
plt.yticks(sorted_time)
plt.savefig('./scatterplot-c5-q1.png')
```
